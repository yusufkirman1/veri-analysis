```python
veri.drop(["number of cases"],axis=1,inplace=True)
veri.drop(["Date"],axis=1,inplace=True)



#yarısından fazlası boş doldurmuyoruz ve drop ediyoruz.
```

```python
veri = veri.interpolate(method='linear',limit_direction='both', limit_area=None)
#burda da bir sonraki veri ile bi önceki verinin ortalaması alınıp dolduruldu
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=veri.drop("number of deaths",axis=1)
y=veri["number of deaths"]
```

```python
x.head(5)
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,
                                              random_state=100)  
 # burda test ve train oranını verdik yüzde 75 i train içinken kalan yüzde 25 test için kullanılıcak
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled=scaler.fit_transform(x_test)
```

```python
lg_model=LogisticRegression(max_iter=500)#deneme amaçlı kullanıldı ancak logistic regreesion 0,1 gibi çıktılar almamız gerektiğinde kullanılır
```

```python
lg_model.fit(x_train_scaled,y_train) #ilk çalıştırdığımıda hata aldım çünkü linear regression içinde default iteration number 100 bunu 500 e çıakrtarak hhatadan kurtuluyoruz
```

```python
lg_model.score(x_test_scaled,y_test)
```

```python
lg_model.score(x_train_scaled,y_train) 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error # Regresyon metrikleri
```

```python
reg_model = LinearRegression()
reg_model.fit(x_train_scaled, y_train)
```

```python
train_score = reg_model.score(x_train_scaled, y_train)
test_score = reg_model.score(x_test_scaled, y_test)
```
